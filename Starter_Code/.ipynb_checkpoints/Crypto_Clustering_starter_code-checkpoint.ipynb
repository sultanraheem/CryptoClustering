```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_scaled_data = pd.DataFrame(scaled_data, columns=df_market_data.columns)

# Copy the crypto names from the original data
df_scaled_data['crypto_name'] = df_market_data.index

# Set the coin_id column as index
df_scaled_data.set_index('crypto_name', inplace=True)

# Display sample data
print("Sample Scaled Data:")
print(df_scaled_data.head(10))
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

#Just to shhow the values as well
print("List of k-values:")
print(k_values)
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # 2. Fit the model to the data using `df_scaled_data`
    kmeans.fit(df_scaled_data)
    
    # 3. Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans.inertia_)

# Show reference
print("Inertia values for different k-values:")
print(inertia_values)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Show reference
print("Elbow Curve DataFrame:")
print(df_elbow)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

# Import required libraries
import matplotlib.pyplot as plt

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Plot a line chart with all the inertia values computed with the different values of k
plt.figure(figsize=(10, 6))
plt.plot(df_elbow['k'], df_elbow['inertia'], marker='o')
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Around 4 it starts to flatten

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
kmeans = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the scaled data
kmeans.fit(df_scaled_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
cluster_labels = kmeans.predict(df_scaled_data)

# Print the resulting array of cluster values
print("Resulting array of cluster values:")
print(cluster_labels)
```

```python
# Create a copy of the DataFrame
df_scaled_data_copy = df_scaled_data.copy()

# Display the first few rows of the copied DataFrame
print("Copied DataFrame:")
print(df_scaled_data_copy.head())
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_scaled_data_copy['Cluster'] = cluster_labels

# Display sample data
print("Sample Data with Cluster Labels:")
print(df_scaled_data_copy.head(10))
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

import hvplot.pandas

scatter_plot = df_scaled_data_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Cluster",
    hover_cols=['coin_id'],
    title="Cryptocurrency Clusters",
    width=800,
    height=400
)

# Display show reference basically
scatter_plot

```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
from sklearn.decomposition import PCA
# With PCA
pca = PCA(n_components=3)
# 3 comps show reference
print(pca)
```

```python
# Import necessary libraries
from sklearn.decomposition import PCA

# Create a PCA model instance and set n_components=3
pca = PCA(n_components=3)

# Fit the PCA model and transform the data to three principal components
pca_data = pca.fit_transform(df_scaled_data)

# Create a DataFrame with the PCA data
df_pca = pd.DataFrame(pca_data, columns=['Principal Component 1', 'Principal Component 2', 'Principal Component 3'])

# Display the first five rows of the DataFrame
print("First five rows of the PCA DataFrame:")
print(df_pca.head())
```

```python
# Import necessary libraries
from sklearn.decomposition import PCA

# Create a PCA model instance and set n_components=3
pca = PCA(n_components=3)

# Fit the PCA model to the data
pca.fit(df_scaled_data)

# Retrieve the explained variance to determine how much information can be attributed to each principal component
explained_variance = pca.explained_variance_ratio_

# Print the explained variance for each principal component
print("Explained variance for each principal component:")
print(explained_variance)
```

```python
# Import necessary libraries
from sklearn.cluster import KMeans
import pandas as pd

# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):
    # Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # Fit the model to the data using df_pca
    kmeans.fit(df_pca)
    
    # Append the model.inertia_ to the inertia list
    inertia_values.append(kmeans.inertia_)

# Display the inertia values
print("Inertia values for different k-values:")
print(inertia_values)
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
import pandas as pd
import numpy as np
import hvplot.pandas
from sklearn.cluster import KMeans

# Example data (replace with your own data or use generated data)
np.random.seed(0)
X = np.random.rand(100, 2)

# Initialize lists to store inertia values for each k
inertia_values_k1 = []
inertia_values_k2 = []
inertia_values_k3 = []
inertia_values_k4 = []
inertia_values_k5 = []
k_values = range(1, 6)  # Example range of k values

# Compute inertia for each k value
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    if k == 1:
        inertia_values_k1.append(kmeans.inertia_)
    elif k == 2:
        inertia_values_k2.append(kmeans.inertia_)
    elif k == 3:
        inertia_values_k3.append(kmeans.inertia_)
    elif k == 4:
        inertia_values_k4.append(kmeans.inertia_)
    elif k == 5:
        inertia_values_k5.append(kmeans.inertia_)

# Create separate line plots for each inertia values using hvplot
plot_k1 = pd.DataFrame(inertia_values_k1, columns=['Inertia']).hvplot.line(xlabel='Number of Clusters (k)', ylabel='Inertia', label='k=1')
plot_k2 = pd.DataFrame(inertia_values_k2, columns=['Inertia']).hvplot.line(xlabel='Number of Clusters (k)', ylabel='Inertia', label='k=2')
plot_k3 = pd.DataFrame(inertia_values_k3, columns=['Inertia']).hvplot.line(xlabel='Number of Clusters (k)', ylabel='Inertia', label='k=3')
plot_k4 = pd.DataFrame(inertia_values_k4, columns=['Inertia']).hvplot.line(xlabel='Number of Clusters (k)', ylabel='Inertia', label='k=4')
plot_k5 = pd.DataFrame(inertia_values_k5, columns=['Inertia']).hvplot.line(xlabel='Number of Clusters (k)', ylabel='Inertia', label='k=5')

# Combine all plots into a composite plot for Elbow curves
composite_plot_elbow = (plot_k1 * plot_k2 * plot_k3 * plot_k4 * plot_k5).opts(
    legend_position='top_left', title='Elbow Curves for Different Values of k')

# Display the composite plot for Elbow curves
composite_plot_elbow
```

```python
import hvplot.pandas

# Create scatter plots for each cluster
scatter_plots = []
for cluster in range(4):  # Assuming there are 4 clusters
    scatter_plot = df_pca_copy[df_pca_copy['Cluster'] == cluster].hvplot.scatter(
        x='Principal Component 1',
        y='Principal Component 2',
        by='Cluster',  # Color points by cluster labels
        hover_cols=['coin_id'],  # Display coin_id on hover
        title=f'Cluster {cluster}',
        width=400,
        height=300,
        cmap='Category10'  # Specify colormap for cluster colors
    )
    scatter_plots.append(scatter_plot)

# Combine all scatter plots into a composite plot for clusters
composite_plot_clusters = scatter_plots[0] + scatter_plots[1] + scatter_plots[2] + scatter_plots[3]

# Display the composite plot for clusters
composite_plot_clusters
```
